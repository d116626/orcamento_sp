```python
df = pd.read_csv('../data/servidores/remuneracao_0420.txt', sep=';' ,encoding = 'ISO-8859-9', decimal=',')
detran = df[df['ORGÃO']=='DEPARTAMENTO ESTADUAL DE TRANSITO DETRAN']

print('Abril de 2020 \n')

print('Total Servidores    = ',len(detran))
print('Remuneração Média   = ',round(detran['REMUNERAÇÃO DO MÊS'].mean()))
print('Remuneração Total   = ',round(detran['REMUNERAÇÃO DO MÊS'].sum()))
print('Remuneração Liquida = ', round(detran['TOTAL LIQUÍDO (R$)'].sum()), '\n\n')


# detran.describe()
```

```python
detran['Remuneração Total'] = detran['REMUNERAÇÃO DO MÊS']
detran['Remuneração Média'] = detran['REMUNERAÇÃO DO MÊS']
detran['Quantidade'] = detran['CARGO']
```

```python
detran_cargos = detran.groupby(by='CARGO', as_index=False).agg({'Quantidade':'count', 'Remuneração Média':'mean', 'Remuneração Total':'sum'}).round().sort_values('Quantidade',ascending=False)
```

```python
detran_cargos.to_excel('../data/servidores/detran_remunecacoes_cargos_abril_2020.xlsx', index=False)
```
