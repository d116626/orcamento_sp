```python
df_20 = pd.read_excel('../2020/orcamento_fiscal_2020.xlsx',encoding='utf-8')
df_19 = pd.read_excel('../2019/orcamento_fiscal_2019.xlsx',encoding='utf-8')
df_18 = pd.read_excel('../2018/orcamento_fiscal_2018.xlsx',encoding='utf-8')
```

```python
# cols = ['orgaos','unidade','programas','acao']
cols = ['orgaos']
refugee_df = df_18.groupby(by = cols, as_index=False).sum()[['orgaos','valor']]
```

```python
refugee_df['orgao'] = refugee_df['orgaos'].str.split('-').apply(lambda x: x[1]).str.strip()
```

```python
refugee_df['source'] = [i for i in range(len(refugee_df))]
```

```python
refugee_df['target'] = max(refugee_df['source'])+1
```

```python
refugee_df['color'] = '#262C46'
```

```python
data_trace = dict(
    type='sankey',
    domain = dict(
      x =  [0,1],
      y =  [0,1]
    ),
    orientation = "h",
    valueformat = ".0f",
    node = dict(
      pad = 10,
      thickness = 30,
      line = dict(
        color = "black",
        width = 0.5
      ),
      label =  refugee_df['orgao'].dropna(axis=0, how='any'),
      color = refugee_df['color']
    ),
    link = dict(
      source = refugee_df['source'].dropna(axis=0, how='any'),
      target = refugee_df['target'].dropna(axis=0, how='any'),
      value = refugee_df['valor'].dropna(axis=0, how='any'),
  )
)

layout =  dict(
    title = "Orçamento do Estado",
    height = 772,
    width = 950,
    font = dict(
      size = 10
    ),    
)


fig = dict(data=[data_trace], layout=layout)
iplot(fig, validate=False)
```
